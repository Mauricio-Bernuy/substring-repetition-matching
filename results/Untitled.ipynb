```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cnt = [5,10,50,100,250,500,750,1000,1500,2000]
```

```python
df = pd.read_csv('substring-repetition-matching/worst_results.csv')
df.index = cnt
```

```python
df
```

```python
for col in df.columns:
    ax = df[col].plot(kind='line', figsize=(20,10))

    title = col.upper()+' (WORST CASE)'
    ax.set_title(title, size=25
                )
    ax.set_ylabel('TIEMPO PROMEDIO(ms)', size=20)
    ax.set_xlabel('LONGITUD DE LA CADENA', size = 20)
    plt.savefig(title+'.png')
    plt.show()

```

```python
ax = df.plot(kind='line', figsize=(20,10))

title = 'COMPARACIN DE TIEMPOS ENTRE LOS ALGORITMOS IMPLEMENTADOS (WORST CASE)'
ax.set_title(title, size=25
            )
ax.set_ylabel('TIEMPO PROMEDIO(ms)', size=20)
ax.set_xlabel('LONGITUD DE LA CADENA', size = 20)
plt.savefig(title+'.png')

```
